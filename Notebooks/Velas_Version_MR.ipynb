Obtenemos el dataset de trabajo leyendo el archivo CSV velas_cryptos

```python
# datos para generar el dataset: bajados de Kaggle : https://www.kaggle.com/rounakbanik/the-movies-dataset
ruta= '../Datos/'
extencion= '.csv'
file= 'velas_cryptos'
```

```python
def lectura_data(ruta, extencion, archivo):
    data= pd.read_csv(ruta+archivo+extencion, low_memory=False)
    return data
```

```python
data= lectura_data(ruta, extencion, file)
```

```python
def plot_crypto(crypto, fecha_desde,_fecha_hasta,data):
    crypto_mask = np.logical_and(np.logical_and(data.crypto == crypto , data.timestamp > fecha_desde ),
                           data.timestamp < fecha_hasta)
    df = data[crypto_mask].copy(deep=True)
    df.index = pd.DatetimeIndex(df.timestamp)
    df = df.rename(columns={'open':'Open','close':'Close','min':'Low','max':'High'})
    mplf.plot(df,type='candle',title='Velas ' +crypto+ ' last 90 days',style='charles')
```

```python
plot_crypto('ADAUSD','2021-01-01','2021-03-31',data)
plot_crypto('BTCUSD','2021-01-01','2021-03-31',data)
plot_crypto('ETHUSD','2021-01-01','2021-03-31',data)
```
