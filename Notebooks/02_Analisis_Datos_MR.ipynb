- Exponente de Hurst
    - Si > 0.5 autocorrelacionado directo
    - Si = 0.5 Movimiento browniano
    - Si < 0.5 autocorrelacionado inverso.... si esta subiendo, va a bajar

```python
def hurst(ts):
    largo =ts.shape[0]
    lags = range(20,min(100,(largo - 1)))
    tau = [sqrt(std(subtract(ts[lag:].reset_index(),ts[:-lag].reset_index()))[1]) for lag in lags]   
    poly = polyfit(log(lags),log(tau),1)    
    return poly[0]**2.0
```

```python

h = hurst(df.close)
print(f'> Hurst Exponent:{h}')
```

```python
def plot_hurst(H,c,data):
    f, ax = plt.subplots()
    ax.plot(data[0], c*data[0]**H, color="deepskyblue")
    ax.scatter(data[0], data[1], color="purple")
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel('Time interval')
    ax.set_ylabel('R/S ratio')
    ax.grid(True)
    plt.show()
    print("H={:.4f}, c={:.4f}".format(H,c))
```

```python
H, c, data = compute_Hc(df.close, kind='change',min_window=20,max_window=100, simplified=True)
plot_hurst(H,c,data)
```

```python
MyHValues = []
hurstDF =pd.DataFrame(columns=['H1'])

Bloque_Datos = 60
Start = 0
Fin = df.shape[0]
End = Bloque_Datos

while Start < Fin:
    hurstDF = df.close[Start:End] 
    MyHValues.append(hurst(hurstDF))
    Start =Start + Bloque_Datos
    End   = End + Bloque_Datos
   

print(f"Media del exponente hurst (H): {(pd.Series(MyHValues).mean())}")
```
